| Column | Description |

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
# defining variables
netflix_movies = netflix_df[netflix_df["type"] == "Movie"]
print(netflix_movies)

# selecting variables for netflix_movies
movies_in_1990s = netflix_movies[(netflix_movies["release_year"] >= 1990) & (netflix_movies["release_year"] < 2000)]
print(movies_in_1990s)

# plotting through histogram
plt.hist(movies_in_1990s["duration"], bins = 20)
plt.xlabel('Duration')
plt.ylabel('Number of Movies')
plt.show()

# count of short action movies
duration = 100
short_movie_count = 0
action_movies = movies_in_1990s[movies_in_1990s["genre"] == "Action"]
for lab, rows in action_movies.iterrows() :
    if rows["duration"] < 90 :
        short_movie_count = short_movie_count + 1
    else :
        short_movie_count = short_movie_count
print(short_movie_count)
    
```
